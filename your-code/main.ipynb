## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
"""
Calculate:
p = probability that the fruit is an apple 
q = probability that the fruit is an orange
"""
````

````python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

total_fruits = 100
apple = 60
orange = 40

# Calculate probabilities
p = apple / total_fruits  # Probability of picking an apple
q = orange / total_fruits  # Probability of picking an orange

print(f"The probability of picking an apple is {p:.1%}")
print(f"The probability of picking an orange is {q:.1%}")

# Create Bernoulli Distribution object
# p = 0.6 (success = picking an apple)
bd = bernoulli(p)

# Possible outcomes for Bernoulli: 0 (orange) or 1 (apple)
x = [0, 1]

# Calculate probabilities using PMF (Probability Mass Function)
probabilities = bd.pmf(x)
print(f"\nBernoulli Distribution probabilities:")
print(f"P(X=0) = P(orange) = {probabilities[0]:.3f}")
print(f"P(X=1) = P(apple) = {probabilities[1]:.3f}")

# Visualization of Bernoulli Distribution
plt.figure(figsize=(10, 6))
plt.bar(x, probabilities, color=['orange', 'red'], alpha=0.7, width=0.4)
plt.title('Bernoulli Distribution (p=0.6)', fontsize=16)
plt.xlabel('Outcome (0=Orange, 1=Apple)', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.xticks([0, 1], ['Orange (0)', 'Apple (1)'])
plt.ylim(0, 0.8) # it sets the limit to 0.8

# Add probability labels on bars
for i, prob in enumerate(probabilities):
    plt.text(i, prob + 0.02, f'{prob:.3f}', ha='center', fontsize=12)

plt.grid(axis='y', alpha=0.3)
plt.show()

# Generate random samples from Bernoulli distribution
random_samples = bernoulli.rvs(p, size=30)
print(f"\nRandom sampling results:")
print(f"Generated {len(random_samples)} random samples")
print(f"Number of apples (1s): {sum(random_samples)}")
print(f"Number of oranges (0s): {len(random_samples) - sum(random_samples)}")
print(f"Empirical probability of apple: {sum(random_samples)/len(random_samples):.3f}")

# Statistical properties of the Bernoulli distribution
print(f"\nBernoulli Distribution Statistics:")
print(f"Mean (expected value): {bd.mean():.3f}")
print(f"Variance: {bd.var():.3f}")
print(f"Standard deviation: {bd.std():.3f}")
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# Probability that the sample contains exactly 5 apples (and thus 15 oranges) in 20 picks
n_trials = 20
k_successes = 5

# Binomial probability mass function
prob_5_apples_15_oranges = stats.binom.pmf(k_successes, n_trials, p)
print(f"Probability of getting exactly 5 apples and 15 oranges in 20 picks: {prob_5_apples_15_oranges:.6f}")

# Binomial distribution: probability for all possible numbers of apples in 20 picks
k_values = np.arange(0, n_trials + 1)
binom_probs = stats.binom.pmf(k_values, n_trials, p)

plt.figure(figsize=(8, 5))
plt.bar(k_values, binom_probs, color='skyblue', alpha=0.7)
plt.title(f'Binomial Distribution (n={n_trials}, p={p})')
plt.xlabel('Number of Apples in 20 Picks')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(axis='y', alpha=0.3)
plt.show()
````

P(X=k)=((2.3**5)*e**(-2.3))5!

````python
# your code here 
import math

# Given values
lam = 2.3
k = 5
e = math.exp(-lam)
fac_k = math.factorial(k) # n is the number you want to find the factorial of
l_elev_k = pow(lam, k)
poisson = (l_elev_k) * (e) / fac_k

print(poisson)

def poisson_pmf(lam, k):
    """
    Calculate the Poisson probability P(X=k) for given lambda and k.
    """
    return (pow(lam, k) * math.exp(-lam)) / math.factorial(k)
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
# Please label the axes and give a title to the plot 

goals = np.arange(0, 11) 
probabilities = [poisson_pmf(lam, k) for k in goals] 

plt.figure(figsize=(10, 6))
plt.bar(goals, probabilities, color='green', alpha=0.7, width=0.6)
plt.title('Poisson Distribution - Goals in UEFA Champions League (lambda=2.3)', fontsize=14)
plt.xlabel('Number of Goals', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.xticks(goals)
plt.grid(axis='y', alpha=0.3)
plt.show()

````
